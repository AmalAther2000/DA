```python
def print_menu():
    print("\n1. Calculate Speed")
    print("2. Calculate Distance")
    print("3. Calculate Time")
    print("4. Exit")

while True:
    print_menu()
    choice = input("Choose (1-4): ").strip()

    if choice == "1":
        distance = float(input("Distance: "))
        time = float(input("Time: "))
        speed = distance / time
        print("Speed =", speed)

    elif choice == "2":
        speed = float(input("Speed: "))
        time = float(input("Time: "))
        distance = speed * time
        print("Distance =", distance)

    elif choice == "3":
        distance = float(input("Distance: "))
        speed = float(input("Speed: "))
        time = distance / speed
        print("Time =", time)

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```

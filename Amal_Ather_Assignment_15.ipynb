#### **1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')**

#### **2. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000**

#### **3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')**

#### **4.Show The Top 10 highest Paid Employees**

#### **5.Show All Employees Ordered By Their TotalPay Benefits In Descending Order**

```python
import sqlite3
import pandas as pd

db_path = 'database.sqlite'
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

task_1_query = """
SELECT JobTitle, AVG(TotalPayBenefits)
FROM Salaries
WHERE JobTitle NOT IN ('Not Provided', 'Not provided')
GROUP BY JobTitle;
"""
task_1_result = pd.read_sql(task_1_query, conn)

task_2_query = """
SELECT JobTitle, OvertimePay
FROM Salaries
WHERE OvertimePay > 50000;
"""
task_2_result = pd.read_sql(task_2_query, conn)

task_3_query = """
SELECT COUNT(DISTINCT EmployeeName)
FROM Salaries
WHERE EmployeeName NOT IN ('Not Provided', 'Not provided');
"""
task_3_result = pd.read_sql(task_3_query, conn)

task_4_query = """
SELECT EmployeeName, TotalPayBenefits
FROM Salaries
ORDER BY TotalPayBenefits DESC
LIMIT 10;
"""
task_4_result = pd.read_sql(task_4_query, conn)
task_5_query = """
SELECT EmployeeName, TotalPayBenefits
FROM Salaries
ORDER BY TotalPayBenefits DESC;
"""
task_5_result = pd.read_sql(task_5_query, conn)

conn.close()
task_1_result.head()
task_2_result.head()
task_3_result.head()
task_4_result.head()
task_5_result.head()
```
